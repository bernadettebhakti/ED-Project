```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('ED.csv')
list(df)
```

```python
df = df[['Multiple Attn in Period Flag',
         'ED Arrival Mode',
         'Patient Sex',
         'Patient Age',
         'ED Category',
         'ED Arrival',
         'In Triage',
         'In Treatment',
         'ED Discharge',
         'Patient Postcode']]
df = df.dropna(axis=0)
```

```python
df.isnull().values.any()
```

```python
df.head()
```

```python
df.loc[(df['Patient Postcode'] == 'NW4 4')&(df['Patient Age'] == 54.0)]
```

```python
date_format = '%d/%m/%Y %H:%M'
df['ED Arrival'] =  pd.to_datetime(df['ED Arrival'], format=date_format)
df['ED Discharge'] =  pd.to_datetime(df['ED Discharge'], format=date_format)
```

```python
df.head()
```

```python
df['los'] = (df['ED Discharge'] - df['ED Arrival'])/ pd.Timedelta(minutes=1)
```

```python
df.head()
```

```python
df.los
```

```python
df.los.describe().apply(lambda x: format(x,'f'))
```

```python
plt.hist(df['los'])
plt.title('Distribution of LOS in ED')
plt.ylabel('Count')
plt.xlabel('Length-of-Stay (days)')
#plt.tick_params(top=False, right=False)
plt.show()
```

```python
df.drop(columns=['In Triage',	'In Treatment',	'ED Discharge'], inplace=True)
```
